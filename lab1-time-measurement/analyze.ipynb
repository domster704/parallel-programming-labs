```python
def analyze(times, p=0.95):
    t = np.array(times)
    K = len(t)
    tmin = np.min(t)
    tmax = np.max(t)

    tavg = np.mean(t)
    D = np.mean((t - tavg) ** 2)
    sigma = np.sqrt(D)

    T1 = t[np.abs(t - tavg) <= 3 * sigma]
    tavg1 = np.mean(T1)
    T2 = np.sort(T1)

    # доверительный интервал
    I = int(len(T2) * p)

    best_start = 0
    best_width = float("inf")

    for i in range(len(T2) - I + 1):
        width = T2[i + I - 1] - T2[i]
        if width < best_width:
            best_width = width
            best_start = i

    interval = T2[best_start:best_start + I]
    tavg3 = np.mean(interval)
    sigma1 = np.sqrt(np.mean((T1 - tavg1) ** 2))
    delta = 1.96 * sigma1 / np.sqrt(len(T1))
    ci_low = tavg1 - delta
    ci_high = tavg1 + delta

    return {
        "t": t,
        "T1": T1,
        "interval": interval,
        "tmin": tmin,
        "tmax": tmax,
        "tavg": tavg,
        "tavg1": tavg1,
        "tavg3": tavg3,
        "delta": delta,
        "ci_low": ci_low,
        "ci_high": ci_high
    }
```

```python
def hist_raw(times, cpu, method):
    t = np.array(times)
    tmin = np.min(t)
    tmax = np.max(t)

    margin = (tmax - tmin) * 0.2

    plt.figure(figsize=(10, 4))

    plt.hist(t, color="red", edgecolor="black")

    plt.xlim(tmin - margin, tmax + margin)

    plt.title(f"{cpu} — {method}\nГистограмма без отсева грубых выбросов")

    plt.xlabel("Время, мс")
    plt.ylabel("Количество измерений")

    plt.grid(alpha=0.3)

    plt.show()
```

```python
def hist_filtered(times, cpu, method):
    res = analyze(times)
    t = res["T1"]
    tmin = np.min(times)
    tmax = np.max(times)

    margin = (tmax - tmin) * 0.2

    plt.figure(figsize=(10, 4))

    plt.hist(t, bins=25, color="red", edgecolor="black")

    plt.xlim(tmin - margin, tmax + margin)

    plt.title(f"{cpu} — {method}\nГистограмма с отсевом грубых выбросов (пример мультимодального распределения)")

    plt.xlabel("Время, мс")
    plt.ylabel("Количество измерений")

    plt.grid(alpha=0.3)

    plt.show()
```

```python
def hist_mode(times, cpu, method):
    res = analyze(times)
    t = res["interval"]
    tmin = np.min(times)
    tmax = np.max(times)

    margin = (tmax - tmin) * 0.2

    plt.figure(figsize=(10, 4))

    plt.hist(t, bins=25, color="red", edgecolor="black")

    plt.xlim(tmin - margin, tmax + margin)

    plt.title(f"{cpu} — {method}\nГистограмма с отсевом грубых выбросов (унимодальное\n распределение, близкое к нормальному)")

    plt.xlabel("Время, мс")
    plt.ylabel("Количество измерений")

    plt.grid(alpha=0.3)

    plt.show()
```

```python
def result_plot(times, cpu, method):
    res = analyze(times)
    t = res["t"]
    tmin = np.min(times)
    tmax = np.max(times)

    margin = (tmax - tmin) * 0.2

    plt.figure(figsize=(10, 5))

    plt.scatter(np.zeros(len(t)), t, s=20, label="ti")

    plt.scatter(1, res["tmin"], s=120, label="tmin")

    plt.errorbar(2, res["tavg"], yerr=np.std(t),
                 fmt='s', capsize=6, label="tavg(T)")

    plt.errorbar(3, res["tavg1"], yerr=np.std(res["T1"]),
                 fmt='s', capsize=6, label="tavg(T')")

    plt.errorbar(4, res["tavg3"], yerr=np.std(res["interval"]),
                 fmt='s', capsize=6, label="tavg(T''')")

    plt.ylim(tmin - margin, tmax + margin)

    plt.xticks(
        [0, 1, 2, 3, 4],
        ["ti", "tmin", "tavg(T)", "tavg(T')", "tavg(T''')"]
    )

    plt.ylabel("t, ms")

    plt.title(f"{cpu} — {method}\nMeasurement results")

    plt.grid(alpha=0.3)

    plt.legend()

    plt.show()
```

```python
def print_results(cpu, method, times):
    res = analyze(times)
    print("\nCPU:", cpu)
    print("Method:", method)
    print("tmin =", round(res["tmin"], 3))
    print("tavg(T) =", round(res["tavg"], 3))
    print("tavg(T') =", round(res["tavg1"], 3))
    print("Δ =", round(res["delta"], 3))
    print(
        "CI = [",
        round(res["ci_low"], 3),
        ";",
        round(res["ci_high"], 3),
        "]"
    )
```

```python
for cpu in data:
    for method in data[cpu]:
        times = data[cpu][method]

        print_results(cpu, method, times)

        hist_raw(times, cpu, method)
        hist_filtered(times, cpu, method)
        hist_mode(times, cpu, method)

        result_plot(times, cpu, method)
```
